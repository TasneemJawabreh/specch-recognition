```python
import os
import pandas as pd

    #  data is a dictionary with file names as keys and associated DataFrames as values
data = {
    '1.wav': df1,
    '2.wav': df2,
    '3.wav': df3,
    '4.wav': df4,
    '5.wav': df5,
    '6.wav': df6,
    '7.wav': df7,
    '8.wav': df8,
    '10.wav': df10,
    '11.wav': df11,
    '12.wav': df12,
    '13.wav': df13,
    '14.wav': df14,
    '15.wav': df15,
    '16.wav': df16
}
training = {'9.wav': df9}


def split_audio(input_file, output_folder, data):
    # Load the audio file
    audio = AudioSegment.from_file(input_file)

    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Retrieve the DataFrame associated with the current audio file
    df = data.get(os.path.basename(input_file), None)

    if df is None:
        print(f"No data found for {os.path.basename(input_file)}")
        return

    # Read start times, end times, and labels from the DataFrame
    with open(os.path.join(output_folder, 'output_info.txt'), 'wb') as file:
        # Write content to the file
        file.write(f"{output_folder}.\n".encode('utf-8'))

        # Split the audio based on specified start and end times
        for index, row in df.iterrows():
            start_ms = int(float(row['start']) * 1000)  # Convert to milliseconds
            end_ms = int(float(row['end']) * 1000)      # Convert to milliseconds
            output_filename = f"chunk_{index + 1}.wav"
            file.write(f"{output_filename}: {row['sentence']}\n".encode('utf-8'))

            # Extract the corresponding segment
            segment = audio[start_ms:end_ms]

            output_path = os.path.join(output_folder, output_filename)

            # Save each segment to the output folder
            segment.export(output_path, format="wav")

def process_audio_folder(audio_folder, output_base_folder, data):
    # Iterate over each audio file in the input folder
    for audio_filename in os.listdir(audio_folder):
        if audio_filename.endswith(".wav"):
            audio_path = os.path.join(audio_folder, audio_filename)

            # Output folder will be created inside the output_base_folder
            output_folder = os.path.join(output_base_folder, os.path.splitext(audio_filename)[0])           
            split_audio(audio_path, output_folder, data)
            
            
audio_folder = "Audios"  
output_base_folder = "Output"  

process_audio_folder(audio_folder, output_base_folder, data)
```

```python
def normalize_spectrogram(spec):
    # Scale values to be in the range of -1 to 1
    spec_min = np.min(spec)
    spec_max = np.max(spec)
    scaled_spec = 2 * (spec - spec_min) / (spec_max - spec_min) - 1

    # Center values around zero
    mean_spec = np.mean(scaled_spec)
    centered_spec = scaled_spec - mean_spec

    return centered_spec
```

```python
import torch
from typing import List, Dict, Union
from transformers import WhisperProcessor

class DataCollatorSpeechSeq2SeqWithPadding:
    def __init__(self, processor):
        self.processor = processor

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        input_features = [{"input_features": feature["audio"]} for feature in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")

        label_features = [{"input_ids": feature["sentence"]} for feature in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")

        attention_mask = labels_batch.attention_mask.unsqueeze(-1).expand_as(labels_batch["input_ids"])
        labels = labels_batch["input_ids"].masked_fill(attention_mask.ne(1), -100)

        if (labels[:, 0] == self.processor.tokenizer.bos_token_id).all().cpu().item():
            labels = labels[:, 1:]

        batch["labels"] = labels

        return batch


processor = WhisperProcessor.from_pretrained(
    "openai/whisper-small", language="arabic", task="transcribe"
)

```

```python
from transformers.models.whisper.english_normalizer import BasicTextNormalizer

normalizer = BasicTextNormalizer()

def compute_metrics(pred):
    pred_ids = pred.predictions
    label_ids = pred.label_ids

    # replace -100 with the pad_token_id
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id

    # we do not want to group tokens when computing the metrics
    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)
    label_str = processor.batch_decode(label_ids, skip_special_tokens=True)

    # compute orthographic wer
    wer_ortho = 100 * metric.compute(predictions=pred_str, references=label_str)

    # compute normalised WER
    pred_str_norm = [normalizer(pred) for pred in pred_str]
    label_str_norm = [normalizer(label) for label in label_str]
    # filtering step to only evaluate the samples that correspond to non-zero references:
    pred_str_norm = [
        pred_str_norm[i] for i in range(len(pred_str_norm)) if len(label_str_norm[i]) > 0
    ]
    label_str_norm = [
        label_str_norm[i]
        for i in range(len(label_str_norm))
        if len(label_str_norm[i]) > 0
    ]
    wer = 100 * metric.compute(predictions=pred_str_norm, references=label_str_norm)

    return {"wer_ortho": wer_ortho, "wer": wer}
```
